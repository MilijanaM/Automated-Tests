```python
import pandas as pd
```

```python
url= 'https://en.wikipedia.org/wiki/Wikipedia:Fundraising_statistics'

data = pd.read_html(url)
```

```python
type(data)
```

```python
type(data[0])
```

```python
len(data)
```

```python
data[0].head()
```

```python
fundrising= data[0]
```

```python
fundrising.dtypes
```

```python
fundrising['Exp']= fundrising['Expenses'].str[2:]
fundrising['Exp']= fundrising['Exp'].str.replace(',', '')
fundrising.head()
```

```python
fundrising['Exp']= pd.to_numeric(fundrising['Exp'])
```

```python
fundrising.dtypes
```

```python
fundrising['Rev']= fundrising['Revenue'].str[2:]
fundrising['Rev']= fundrising['Rev'].str.replace(',', '')
fundrising.dtypes
```

```python
fundrising.head()
```

```python
fundrising.loc[0, 'Rev']= 'spam'
```

```python
fundrising.head()
```

```python
fundrising['Rev']= pd.to_numeric(fundrising['Rev'], errors='coerce')
fundrising.head()
```
